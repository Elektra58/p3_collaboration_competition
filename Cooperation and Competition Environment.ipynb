## Random Actions

```python
n_episodes = 5  # play the game for n episodes

for i_episode in range(1, n_episodes + 1):
    env_info = config.env.reset(train_mode=True)[config.brain_name]

    states = env_info.vector_observations
    print(f'States is a {type(states)} of shape {states.shape}\n', states)
    
    score = np.zeros(config.n_agents)
    print(f'Scores is a {type(score)} of shape {score.shape}\n', score)
    
    max_t = 0

    for t in range(config.max_t):
        max_t = t
        
        actions = np.random.randn(config.n_agents, config.action_dim)
        print(f'Actions is a {type(actions)} of shape {actions.shape}\n', actions)

        env_info = config.env.step(actions)[config.brain_name]

        next_states = env_info.vector_observations
        print(f'Next States is a {type(next_states)} of shape {next_states.shape}\n', next_states)

        rewards = env_info.rewards
        print(f'Rewards is a {type(rewards)} of length {len(rewards)}\n', rewards)

        dones = env_info.local_done
        print(f'Dones is a {type(dones)} of length {len(dones)}\n', dones)


        states = next_states

        score += rewards
        print(f'Score is a {type(score)} of shape {score.shape}\n', score)

        if np.any(dones):
            max_t = t
            break

    print(f'Max score over both agents for episode {i_episode} is {np.max(score)} in {max_t} timesteps')
        
config.env.close()
```
